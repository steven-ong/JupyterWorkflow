## 1. Visualization technique
The techniques I plan to use are: <br>
(1) Simple line and bar chart.<br>
(2) Combination charts (a combined bar and line plot).<br>

Purpose of starting the discussion with a simple line and bar chart followed by a combination of charts (combined bar and line plot) is to show the limitation of using merely a separate line and bar chart vs using a combination of charts.  

Charts in general are supposed to help drive home key points. In this aspect I will discuss in my article my considerations in choosing: <br>
(i) the type of visualisation to use, <br>
(ii) the limitation of dual axis combination charts, and <br>
(iii) alternative charts to complement the dual axis combination charts. <br>

<font color=green> Details about this technique:  </font><br>

Both bar plots and line plots are excellent techniques to use when displaying and educating the reader on trends across time, different products and companies. <br> 

With bar charts, it's easier to see changes when they are larger. Bar graphs are easy to understand, widely used, and can show changes over time. It is the first type of chart many of us learn in elementary school. That gives bar charts an advantage over other graphs that are difficult to interpret or can only show a single data set.

When smaller changes exist, line graphs do better in showing the differences when compared to bar graphs. The small percentage % change can easily be seen with a line chart. Line graphs are also useful to compare changes over the same period of time for more than one group.

A dual axis chart is a great way to easily illustrate the relationship between two different variables. They illustrate a lot of information with limited space and allow you to discover trends you may have otherwise missed if you're switching between graphs. 

However, one must note that combination of bar and line plots are ***not*** good for showing distrubutions, means, medians, modes, or outliers. When that type of analysis is more insightful or requested for by the stakeholders, there are better techniques for visualization which include creating box and whisker plots, and/or violin plots which I hope to be able to convince you the reader, at end of my article. <br>


## 2. Visualization Library
The library I'm using is **Plotly**.


<font color=green>Why I Chose this Package:</font><br>

I like Plotly simply for its aesthetics. I spent the weekend researching and trying out Plotly. I have been mind blown by its capabilities. I find it to have nice features that is easy to understand and implement when compared to Altair which in comparison has sub par documentation to follow. Developing fancy web apps comes with the opportunity cost of time to build and a steep learning curve to start. This is the sweet spot that Plotly has over other tools, which makes visualising data at any scale much easier.<br>

A major advantage of using Plotly is that it encourages us to be as creative as possible with visualisations since any complex plots only use three main concepts: (i) data, (ii) layout, and (iii) figure objects.<br> 

The Plotly JSON visualization schema is a declarative [format](https://medium.com/plotly/introducing-plotly-py-3-0-0-7bb1333f69c6) for creating and sharing interactive, scientific charts. This means that every aspect of the charts (the colors of the bars, the axes grid lines, the position of annotations) are completely described through a set of attributes.  Because it is [declarative](https://en.wikipedia.org/wiki/Declarative_programming) in nature, data scientists can be free from the implementation details. To me the advantage of creating charts as JSON is portability - charts can be shared across any language or platform that understands the schema.

Plotty has good documentation which is easy to understand. There are definitely other tools out there in the marketplace like Matplotlib, Seaborn, Bokeh if you want to build a visualization dashboard like those popularized by New York Times web pages and The Economist. The open source graphing library covers a wide range of graph types, is easy to use and understand when compared to Seaborn, Bokeh and Altair and it is powerful in generating insights<br>

I have always wanted to create bars and lines together in a chart like how I can usually do it via tableau in my day job and I am glad to have to have found a solution to it which I learned and mastered over four days! <br>

With the above said, I do find some limitations with the Plotly Library since it is relatively new. It doesn't have many options when it comes to adjusting the appearance of the graph.  In Plotly Express, the color scheme a graphic uses needs to be included in the creation of every single plot. Seaborn automatically assigns different colors to different categories, whereas Plotly Express does not. This is not a deal breaker for me as the main goal of Plotly Express was to allow users to quickly and easily create interactive visuals while performing exploratory data analytics.

I wanted to look at the relationships between the rating and price, and will include the wine name as a hover label

```python
def basic_scatter():

    fig = px.scatter(df,
                x='WineRating',
                y='VintagePrice',
                color = 'Region', 
                hover_name = 'WineName', 
                title='Wine Rating vs. Price',
                height = 600, 
                width = 1000 

                )   
    fig.show()
    
    return

basic_scatter()
```

You can isolate certain information. One of the coolest features available with Plotly Express is the ability to add animation frames which thereby allows you to view how something changes over a certain variable, and isolate certain information.
```python
## Lets build an histogram
def basic_histogram():

    fig = px.histogram(df, x='WineRating', 
                   title='Rating distribution', 
                   height = 400, 
                   width = 800 
                      
                  )
    fig.show()
    
    return

basic_histogram()
```

From the histogram we can deduce that the rating of wines follows a somewhat bell shape normal distribution. There are only a few wines given the rating of 4.6 and above, with most wines falling in the 4.1 and 4.2 region.  

```python
## By generating a heat map, we can also decode that we only have a few wine names with ratings of more than 4.5 
## (depicted as yellow) as compared to other colors. 
def basic_heat_map():

    fig = px.imshow(df[['WineRating']])

    fig.show()
    
    return

basic_heat_map()
```

Now that we understand the data better, I will like to see how each **region's** wine varies from another. I have this preperception that the average rating / price differs drastically from one region to another. </br>

I like to keep things simple which means working with concise group of data and removing all unnecessary distraction. 

```python
## But first we have to prepare the data for analysis
## Data Wrangling process

wine_region_mean = df.groupby('Region').mean()

df2=wine_region_mean['VintageRating']
df3 = wine_region_mean['VintagePrice']

## building a new dataframe for analysis
df4 = pd.concat([df2,df3], axis=1)
df4 = df4.reset_index()

df4

## df2 = df2.sort_values(ascending = False)
##wine_mean = df.sort_values('VintageRating')
```

```python
# Rating Means by region
## lets start off building charts using Plotly with our favorite bar charts

## Plotly is very intuitive, with the dataframe set up, you can instruct plotly to build a simple chart with a few code lines
def px_bar_chart():

    fig = px.bar(df4,
            y = 'VintageRating',
            x = 'Region', 
            labels = {
                "VintageRating": "Wine Rating"
            },
             
            text_auto='.2s', 
            ##y = 'VintageRating',
            ##hover_name = 'Winery', 
            title = ' Average Wine Rating by Region', 
            color = None,
            height = 600, 
            width = 1000,
            ## color_discrete_Sequence changes the line color  
            color_discrete_sequence=["#acc8d4"]            
            )

    ## To change range of the y axis, I wanted to magnify the effect, so I started off the y-axis with 4.0 as compared to 0.0
    fig.update_yaxes(range=[4.0, 4.4])

    fig.update_xaxes(tickangle=45, tickfont=dict(
                               family='Rockwell', 
                               color='#4a4a4a', 
                               size=12))

    ## To include text in the bars itself
    fig.update_traces(textfont_size=12, textangle=0, textposition="inside", cliponaxis=False)

    fig.show()
    
    return

px_bar_chart()
```

Ok... This is beauty in manimalism.From one glance I can deduce that the average Chilean wine ratings are higher than wine from the popular Burgundy, Napa Valley and Californian regions.  

```python
## Creating scatter plot
def px_line_chart():    
    ### just like how we done it for the bar chart
    fig = px.line(df4,
            y = 'VintagePrice',
            x = 'Region',  
            labels = {
                "VintagePrice": "Wine Price (US$)"
            },  
            title = ' Average Wine Price by Region', 
            color = None,
            height = 600, 
            width = 1000,
            ## color_discrete_Sequence changes the line color
            color_discrete_sequence=["#244747"]              
            )

    fig.update_yaxes(range=[0.0, 250.0])

    ## To change font type, size, color of axis 
    fig.update_xaxes(tickangle=45, tickfont=dict(
                               family='Rockwell', 
                               color='#91b8bd', 
                               size=12)) 
    fig.show()
    
px_line_chart()    
```

With the above scatter plot, I can deduce that the mean Bordeaux wine price is higher than wine price in Napa Valley, Chilean and Californian. 

Whats next? <br>

The two charts above are aesthetically nice to look at. However it is hard to see the co-relation and make any informed comparison. An investor rule of thumb - "Price is what you pay, value is what you get". What this means is that for the price I pay how much is the value derived from it? It is definitely hard to deduce the above with two separate line and bar chart.  

A dual axis chart bridge the gap by combining the line and bar chart together using Wine Rating as the left Y-axis and Wine Price as the second (right) Y-axis. Dual axis chart is a great way to easily illustrate the relationship between two different variables - Wine Price Vs Wine Rating. They illustrate a lot of information with limited space and allow you to discover trends you may have otherwise missed if you're switching between graphs. 


```python
hover_text = []
for index, row in df.iterrows():
    hover_text.append(('<b>{Region}</b><br><br>'+                                        # The HTML <b> tag will make the city name bold. The <br> is used to produce a line break after the city name  
                      'Wine Price: {VintagePrice}<br>'+
                      'Wine Rating: {VintageRating}<br>' 
                       ).format(
                       Region=row['Region'],
                       VintagePrice=row['VintagePrice'],
                       VintageRating=row['VintageRating'],
                                             ))

df['text'] = hover_text
```

```python
# We start off with creating figure with secondary y-axis


fig = go.Figure()

fig.add_trace(                               #Add a bar graph to the figure
        go.Bar(
        x=df4['Region'],                     ## x_list = df4['Region'].tolist()


        ##text=df['text'],
        y=df4['VintageRating'],              ## y1_list = df4['VintagePrice'].tolist(), ## y2_list = df4['VintageRating'].tolist()
        name="Vintage Rating",
        marker_color='#acc8d4',              #Specify the color of the bars    
        ## hoverinfo='text',                    #Hide the hoverinfo

        ),
        )                                    

fig.add_trace(                               #Add the second trace (line graph) to the figure
        go.Scatter(                          
        x=df4['Region'],
        y=df4['VintagePrice'],
        name="Vintage Price",
        mode='lines',                            #Use go.Scatter() and specify the mode to 'lines' to add the line graph
        ## text=df['text'],               
        hoverinfo='text',                        #Pass the 'text' column to the hoverinfo parameter to customize the tooltip
        line = dict(color='#336666', width=3),   #Specify the color of the line
        yaxis="y2" ),                            #By specifying yaxis='y2' we tell plotly that this line chart uses a secondary y-axis
        )                   

        # fig = make_subplots(specs=[[{"secondary_y": True}]])

```

```python
# Create axis objects

fig.update_layout(
      xaxis=dict(
          domain=[0, 1]                 #Sets the domain of this axis (in plot fraction). Play with the numbers to see how it affects the plot
      ),
    yaxis=dict(                         
        title="Wine Rating",               #Add an axis title for the primary axis
        titlefont=dict(
            color="#acc8d4"               #Make the color of the y-axis the same as the bar color
        ),
        tickmode ='array',
        ticktext = df4['VintageRating'],
        tickfont=dict(
            color="#acc8d4"           
        ),
        range = [4, 4.35]
    ),
    yaxis2=dict(
        title="Wine Price (US$)",  #Add axis title for the second y-axis
        titlefont=dict(
            color="#336666"              #Make the color of the second y-axis the same as its corresponding line color
        ),
        tickmode ='array',
        ticktext = df4['VintagePrice'],
        tickfont=dict(
            color="#336666"
        ),
        anchor="x",                   #If set to "free", this axis' position is determined by 'position'.
        overlaying="y",
        side="right",
        
        range=[0, 500]

    ),
    )

##fig.update_yaxes(range=[4.0, 4.4])


fig.show()


```

It looks like we have already gotten a basic dual axis line charts out. However, we can further improve the aesthetics of this visulisation by the following enhancements:

- Including a proper title
- Rescale the chart 
- rescale the x and y axis
- Include hovering text to enhance users' experience with the end product. 

```python
#Fine-tune the chart
fig.update_layout(hoverlabel_bgcolor='#efe8d1',               #Change the background color of the tooltip to light blue
             autosize=False,
             width=1000,
             height=600,
             title_text="Wine Prices Vs Wine Ratings", #Add a chart title
             title_font_family="Times New Roman",
             title_font_size = 20,
             title_font_color="#336666",                     #Specify font color of the title
             title_x=0.5,                                    #Specify the title position
             xaxis=dict(
                    tickfont_size=10,
                    tickangle = 45,
                    showgrid = True,
                    zeroline = True,
                    showline = True,
                    showticklabels = True,
                    tickformat="%b\n%Y"
                    ),
             legend = dict(orientation = 'h', xanchor = "center", x = 0.1, y= 1.11), #Adjust legend position
            
                 )

fig.show()
```

From this chart we can easily decode the following: <br>

- If wine rating and wine prices are our considerations in our purchase of wine, perhaps Napa Valley Wines which are lower in price and higher in rating brings more bang for bucks when compared to the rest of the regions. <br>

- We are paying too much for the wine in Burgundy which is lower in rating but highest in average pricing. That said, the spike in pricing may be due to a couple of high value wine which are produced in this region and require further investigation. 

This is the benefit of the duo y axis visualisation techinique where we can interpret and decode information which will be somewhat difficult to do if the line wine rating and wine price are plotted on two separate chart.  

**Limitation of Dual Axis Charts**

I want to emphasise that the Burgundy and Bordeaux wine are more expensive when compared to the chilean and Argentinian wines which have higher average wine rating. Maybe this finding is not so obvious with the above dual y-axis chart. That will be the limitation for dual y-axis chart where the user and reader are able to identify trendlines across regions, time and products, but it might not be the best visualisation tool to emphasise on data points which stick out like sore thumbs.  

To drive home that point, I was looking through charts in New York Times and The Economist, that was when I had a Eureka moment, falling back on the trustworthy scatter plot. 

```python
def scatter_plot_():

    fig = px.scatter(df4, x='VintagePrice', 
                    y = 'VintageRating',
                    labels = {
                        "VintagePrice": "Wine Price (US$)",
                        "VintageRating": "Wine Rating"
                
                    }, 
                   color = 'Region',
                   size = 'VintagePrice', size_max = 40,
                   title = 'Vintage Rating vs Vintage Price', 
                   color_discrete_sequence = px.colors.colorbrewer.Paired, 
                   hover_name = 'Region'
                    )
    fig.show()
    
    return

scatter_plot_()
```

At one glance, I am able to deduce from this chart that Burgundy wine has a significantly higher price with lower wine rating. 

While it technically took six lines to create this, it still only took a single command.  <br>

In creating a Plotly Express visualization, everything can be done in the same command, from adjusting the size of the graphic, to the colors it uses, to the axes labels. <br>

Plotly Express to me is the easiest way to create and modify a visualization. Also, the visualization is automatically interactive, which brings me to my next point.<br>
